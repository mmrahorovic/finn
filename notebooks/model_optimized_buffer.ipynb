```python
from finn.custom_op.registry import getCustomOp
from finn.core.modelwrapper import ModelWrapper
import numpy as np

model = ModelWrapper("/workspace/results/syn_optimized_buffer/df_model_fifos.onnx")

uram_total = 0
bram_total = 0
total_string=[]
for n in model.graph.node:
    if n.op_type=="StreamingFIFO":
        depth = getCustomOp(n).get_nodeattr("depth")
        uram = np.ceil(depth/4096)
        width = getCustomOp(n).get_nodeattr("folded_shape")[-1]
        if width>1 and depth>=4096:
            print("{}\t{}\t{}\t{}\t{}".format(depth, width, uram,"URAM", n.name))
            uram_total += uram
            ram_style = "ultra"
        elif width==1 and depth==16384:
            print("{}\t{}\t{}\t{}\t{}".format(depth, width, uram,"URAM", n.name))
            uram_total += uram
            ram_style = "ultra"
        elif width==8 and depth==2048:
            print("{}\t{}\t{}\t{}\t{}".format(depth, width, uram,"URAM", n.name))
            uram_total += uram
            ram_style = "ultra"
        else:
            ram_style = "block"
            if width*4==4:
                #bram_cap = 8192
                bram_cap = 4096
            elif width*4==8:
                bram_cap = 4096
            elif width*4==16:
                bram_cap = 2048
            elif width*4==32:
                bram_cap = 1024
            elif width*4==64:
                bram_cap = 512
            bram = np.ceil(depth/bram_cap)
            print("{}\t{}\t{}\t{}\t{}".format(depth, width, bram," BRAM", n.name))
            bram_total += bram 
        
        string = '"'
        string += n.name
        string += '" : {"depth" : '
        string += str(depth)
        string += ', "ram_style" : '
        string += '"'
        string += ram_style
        string += '"'
        string += ', "impl_style" : "vivado"},'
        total_string.append(string)

for el in total_string:
    print(el)


#print("----------------------------------------")
#for n in model.graph.node:
#    if n.op_type=="StreamingFIFO":
#        depth = getCustomOp(n).get_nodeattr("depth")
#        width = getCustomOp(n).get_nodeattr("folded_shape")[-1]
#        if width*4==4:
#            #bram_cap = 8192
#            bram_cap = 4096
#        elif width*4==8:
#            bram_cap = 4096
#        elif width*4==16:
#            bram_cap = 2048
#        elif width*4==32:
#            bram_cap = 1024
#        elif width*4==64:
#            bram_cap = 512
#        bram = np.ceil(depth/bram_cap)
#        if width==1 or depth<4096:
#            if depth==16384:
#                continue
#            print("{}\t{}\t{}".format(depth, width, bram))
#            bram_total += bram
        
print("Total BRAM: {}".format(bram_total))
print("Total URAM: {}".format(uram_total))
```
